## 1. Narysuj wykres podobny do tego ze slajdu nr. 15 z wykadu. Uwzgldnij opisy osi, zakresy wartoci, kolory, znaczniki wartoci, legend i tytuy osi oraz wykresw.

```python
x = np.linspace(0, 10*np.pi, 80)
y1 = lambda x: x ** 1.3 * np.sin(x)
y2 = lambda x: x * np.cos(x)
plt.figure(figsize=(10, 5))
plt.plot(x, y1(x), 'r--x', label='x**1.3*sin(x)')
plt.plot(x, y2(x), 'g:o', label='x*cos(x)')
plt.xticks(np.arange(0, 10*np.pi, np.pi), ['zero', 'pi', '2pi', '3pi', '4pi', '5pi', '6pi', '7pi', '8pi', '9pi', '10pi'])
plt.yticks([-20, 10, 50], [-20, 10, '5*10'])
plt.xlabel('Argumenty')
plt.ylabel('Wartoci')
plt.title('Funkcje')
plt.legend(loc='lower left')
plt.axhline(y=-20, color='k', linestyle='--')
plt.axhline(y=50, color='k', linestyle='--')
```

## 2. Narysuj wykres 10*sin(x)+normal(0,1) dla 1000 punktw w zakresie 0,10. Przyda si funkcja normal() do losowania liczb z rozkadu normalnego.

```python
x = np.linspace(0, 10, 1000)
y = 10 * np.sin(x) + np.random.normal(0, 1, 1000)
plt.figure(figsize=(10,5))
plt.plot(x, y, 'r')
```

## 3. Napisz program, ktry wykona 1000 prb rzutu 10 kostkami. Narysuj na jednym wykresie 3 przenikajce si histogramy (alpha=0.25): suma oczek na kociach parzystych (0,2,4,6,8), suma oczek na kociach nieparzystych (1,3,5,7,9), suma oczek na wszystkich kociach. Moe przyda si funkcja randint()  z moduu random do losowania liczb cakowitych z zakresu [1,6].

```python
x = np.random.randint(1,7, (1000, 10))
sum_even = np.sum(x[:, np.arange(10) % 2 == 0], axis=1)
sum_odd = np.sum(x[:, np.arange(10) % 2 == 1], axis=1)
sum_all = sum_even + sum_odd
```

```python
plt.hist(sum_even, alpha=0.25, label='sum of even')
plt.hist(sum_odd, alpha=0.25, label='sum of odd')
plt.hist(sum_all, alpha=0.25, label='sum of all')
plt.legend()
```

## 4. Przedstaw dane z wykresu 3. w postaci 3 wykresw pudekowych (boxplot)

```python
data = {
    'even': sum_even,
    'odd': sum_odd,
    'all': sum_all
}

plt.boxplot(data.values(), labels=data.keys());
```

## 5. Narysuj powierzchni 3d (x,y,z) dla punktw na siatce linspace(0,1,100) gdzie z=y*cos(x)+(1-y)*sin(x), Moe by przydatny ten przykad

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)

xx, yy = np.meshgrid(x, y)

z_func = lambda x,y: y * np.cos(x) + (1 - y) * np.sin(x)

z = z_func(xx,yy)

surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm, antialiased=False)

ax.set_zlim(z.min(), z.max())
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
