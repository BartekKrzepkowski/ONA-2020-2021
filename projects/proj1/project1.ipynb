```python
from project1 import init_plane_numpy, gen, create_map, create_surface
```

### wersja podstawowa

```python
# całkowita głębokość
N = 12
# inicjowanie tablicy
plane = init_plane_numpy(N=N)

center = len(plane) // 2
init_centers = [[center, center]]
# generuj krajobraz
plane = gen(plane, init_centers, k=N, step=2**(N-1), sigma=0.66, connect=False)
```

```python
# wyświetl mapę
create_map(plane, map_file=None, colormap='terrain')
```

```python
# wyświetl krajobraz 3d
create_surface(plane, N, surf_file=None, colormap='terrain')
```

### wersja nieco bardziej zaawansowana

```python
from project1 import create_centers
```

```python
# całkowita głębokość
N = 12
# głębokość początkowa
l = 7
# inicjowanie tablicy
plane = init_plane_numpy(N=N)

center = len(plane) // 2
init_centers = [[center, center]]
# generuj krajobraz do kroku l
plane = gen(plane, init_centers, k=l, step=2**(N-1), sigma=0.66, connect=True)
```

```python
# wyświetl mapę
create_map(plane, map_file=None, colormap='terrain')
```

```python
# wyświetl krajobraz 3d
create_surface(plane, N, surf_file=None, colormap='terrain')
```

```python
# zbierz centra z ostatniego etapu wczesniejszej płaszczyzny
init_centers = create_centers(init_centers, k=l, step=2**(N-2))
# dokończ generowanie krajobrazu (wzięta jest płaszczyzna wygenerowana wczesniej, ale zamiast niej może być wczytana macierz)
plane = gen(plane, init_centers, k=N-l, step=2**(N-1-l), sigma=0.66, connect=True)
```

```python
# wyświetl mapę
create_map(plane, map_file=None, colormap='terrain')
```

```python
# wyświetl krajobraz 3d
create_surface(plane, N, surf_file=None, colormap='terrain')
```
